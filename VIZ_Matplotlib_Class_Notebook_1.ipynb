##**A Simple Example** 

Matplotlib graphs your data on Figures (i.e., windows, Jupyter widgets, etc.), each of which can contain one or more Axes (i.e., an area where points can be specified in terms of x-y coordinates (or theta-r in a polar plot, or x-y-z in a 3D plot, etc.). The most simple way of creating a figure with an axes is using pyplot.subplots. We can then use Axes.plot to draw some data on the axes:

````
import matplotlib.pyplot as plt #convention
import numpy as np

fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]);  # Plot some data on the axes.
````

Many other plotting libraries or languages do not require you to explicitly create an axes. For example, in MATLAB, one can just do:


```
plot([1, 2, 3, 4], [1, 4, 2, 3])  % MATLAB plot.
```

and get the desired graph.

In fact, you can do the same in Matplotlib: for each Axes graphing method, there is a corresponding function in the matplotlib.pyplot module that performs that plot on the "current" axes, creating that axes (and its parent figure) if they don't exist yet. So the previous example can be written more shortly as


````
plt.plot([1, 2, 3, 4], [1, 4, 2, 3]);  # Matplotlib plot.
````

## **Parts of a Figure**

### **Figure**

The easiest way to create a new figure is with pyplot. It's convenient to create the axes together with the figure, but you can also add axes later on, allowing for more complex axes layouts.

````
fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()  # a figure with a single Axes
fig, axs = plt.subplots(2, 2);  # a figure with a 2x2 grid of Axes
````

### **Axes and Axis**

Axes is where we plot data. You can see that by default it adapts axis to provided data upper and lower limits. You can also see how it assigns different colours automatically.

````
##OO Approach
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x)  # Plot some data on the axes.
ax.plot(x, x**2)  # Plot more data on the axes...
ax.plot(x, x**3);  # ... and some more.
````

````
##Pyplot Approach
x = np.linspace(0, 2, 100)

plt.plot(x, x)  # Plot some data on the axes.
plt.plot(x, x**2)  # Plot more data on the axes...
plt.plot(x, x**3);  # ... and some more.
````

###**Bar**


Simple bar plot:

````
x = ['G1', 'G2', 'G3', 'G4', 'G5', 'G6']
y = [5, 6, 15, 22, 24, 8]

plt.bar(x, y, color='r')
plt.xticks(np.arange(6), x)
plt.xlabel("Groups")
plt.ylabel("Value")
plt.title("Simple bar plot");
````

Add error bar:

````
x = ['G1', 'G2', 'G3', 'G4', 'G5', 'G6']
y = [5, 6, 15, 22, 24, 8]
std_deviation = [1, 2, 7, 4, 2, 3]

plt.bar(x, y, color='r', yerr=std_deviation)
plt.xticks(np.arange(6), x)
plt.xlabel("Groups")
plt.ylabel("Value")
plt.title("Simple bar plot");
````

Horizontal view:

````
x = ['G1', 'G2', 'G3', 'G4', 'G5', 'G6']
y = [5, 6, 15, 22, 24, 8]
std_deviation = [1, 2, 7, 4, 2, 3]

plt.barh(x, y, color='r', xerr=std_deviation)
plt.yticks(np.arange(6), x)
plt.ylabel("Groups")
plt.xlabel("Value")
plt.title("Simple bar plot");
````

Several Xs series:

````
N = 5
a = (20, 35, 30, 35, 27)
b = (25, 32, 34, 20, 25)

ind = np.arange(N) 
width = 0.35       
plt.bar(ind, a, width, label='A')
plt.bar(ind + width, b, width, label='B')

plt.ylabel('Value')
plt.title('Multiples Xs bar plot')

plt.xticks(ind + width / 2, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.legend();
````

Stacked bar plot:

````
N = 5
a = (20, 35, 30, 35, 27)
b = (25, 32, 34, 20, 25)

ind = np.arange(N) 
width = 0.35       
plt.bar(ind, a, width, label='A')
plt.bar(ind, b, width, label='B',bottom=a)

plt.ylabel('Value')
plt.title('Stacked bar plot')

plt.xticks(ind + width / 2, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.legend();
````
